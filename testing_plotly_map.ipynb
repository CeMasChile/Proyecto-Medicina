```python
import geopandas as gpd
import plotly.graph_objects as go
cuarteles = gpd.read_file('./data/Carabineros/Cuarteles_Carabineros/CUARTELES.shp')
bomberos = gpd.read_file('./data/Bomberos/Companias-de-Bomberos/COMPANIAS_DECHILE.shp')
PDI = gpd.read_file('./data/PDI/PDI/UNIDADES_OPERATIVAS_PDI_2017.shp')

#coord sys
PDI = PDI.to_crs(cuarteles.crs)
bomberos = bomberos.to_crs(cuarteles.crs)
```

```python
token=open("token.txt").read()
```

```python
def map_from_df(df, hover, name, color, ms):
    '''
    df: dataframe o geodf
    hover: column to be showed when hovermouse
    name: name to show in the legend and in the text
    color: color of the points
    ms: markersixe
    returns: go object
    '''
    GO= go.Scattermapbox(
    lat=df['geometry'].y,
    lon=df['geometry'].x,
    mode='markers',
    marker=go.scattermapbox.Marker(
        size=ms,
        color=color
    ),
    text=[name],
    hovertext=df[hover],
    name=name
)
    return GO
```

```python
ms=6 #markersize
bomberos_map = map_from_df(bomberos,'Compa√±ia','Bomberos','blue',ms)
cuarteles_map = map_from_df(cuarteles,'NOMBRE_UNI','Cuarteles de Carabineros','green',ms)
PDI_map = map_from_df(PDI,'NOM_UNID','PDI','yellow',ms)

fig = go.Figure([bomberos_map, cuarteles_map, PDI_map])

fig.update_layout(
    hovermode='closest',
    mapbox=go.layout.Mapbox(
        accesstoken=token,
        bearing=0,
        center=go.layout.mapbox.Center(
            lat=-33.4376377,
            lon=-70.6424733
        ),
        pitch=0,
        zoom=10
    )
)
fig.show()
```
